```python
import mysql.connector
```

```python
mydb=mysql.connector.connect(
      host = "localhost",
      user = "root",
      password ="",
      database ="ems"
)
def insert_emp(emp_id,emp_name,emp_salary,emp_dept,email):
    cursor=mydb.cursor()
    query="INSERT INTO EMPLOYEE(emp_id,emp_name,emp_salary,emp_dept,email) VALUES (%s ,%s , %s,%s,%s)"
    values=(emp_id,emp_name,emp_salary,emp_dept,email)
    cursor.execute(query,values)
    result=cursor.fetchall()
    mydb.commit()
    print("Employee added sucessfully")
    cursor.close()
def update_emp(emp_id,emp_salary):
    cursor = mydb.cursor()
    query="UPDATE EMPLOYEE SET EMP_DEPT='IT' WHERE EMP_ID= %s OR EMP_SALARY=%s"
    value=(emp_id,emp_salary)
    cursor.execute(query,value)
    mydb.commit()
    result=cursor.fetchall()
    print("Record affected")
    cursor.close()
def delete_emp(emp_id,emp_name):
    cursor = mydb.cursor()
    query=" DELETE FROM EMPLOYEE WHERE EMP_ID= %s AND EMP_NAME= %s"
    value=(emp_id,emp_name)
    cursor.execute(query,value)
    mydb.commit()
    result=cursor.fetchall()
    print("Record affected")
    cursor.close()
def list_employee():
    cursor = mydb.cursor()
    query=" SELECT * FROM  EMPLOYEE"
    cursor.execute(query)
    rows=cursor.fetchall()
    for row in rows:
        print(row)
    cursor.close()
def sort_employee():
    cursor = mydb.cursor()
    query=" SELECT * FROM  EMPLOYEE ORDER BY EMP_DEPT AND EMP_SALARY"
    cursor.execute(query)
    rows=cursor.fetchall()
    for row in rows:
        print(row)
    cursor.close()
def sort_employee():
    cursor = mydb.cursor()
    query=" SELECT * FROM  EMPLOYEE ORDER BY EMP_SALARY "
    cursor.execute(query)
    rows=cursor.fetchall()
    for row in rows:
        print(row)
    cursor.close()
def filter_employee():
    cursor = mydb.cursor()
    query=" SELECT * FROM EMPLOYEE WHERE EMP_DEPT='IT' "
    cursor.execute(query)
    rows=cursor.fetchall()
    for row in rows:
        print(row)
    cursor.close()
if __name__=="__main__":
    while True:
        print("CRUD FUNCTIONALITY")
        print("1. INSERT")
        print("2. UPDATE")
        print("3. DELETE")
        print("4. LIST")
        print("5. SORT")
        print("6. FILTER")
        ch=int(input("Enter your choice"))
        if ch==1:
            emp_id=input("Enter ID ")
            emp_name=input("Enter the employee name ")
            emp_salary=float(input("Enter employee salary "))
            emp_dept=input("Enter employee department  ")
            email=input("Enter email address ")
            insert_emp(emp_id,emp_name,emp_salary,emp_dept,email)
        elif ch==2:
            emp_id=input("Enter the employee id ")
            emp_salary=float(input("Enter employee salary "))
            update_emp(emp_id,emp_salary)
        elif ch==3:
            emp_id=input("Enter the id ")
            emp_name=input("Enter the employee name ")
            delete_emp(emp_id,emp_name)
        elif ch==4:
            list_employee()
        elif ch==5:
            sort_employee()
        elif ch==6:
            filter_employee()
            
        else:
            print("Invalid choice Pls try again")
mydb.close()
```
